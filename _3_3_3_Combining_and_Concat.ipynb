### Concatenating

```python
bank = pd.read_csv("/content/Practicum_Datasets/bank.csv", delimiter=';')
bank
```

```python
bank.info()
```

#### Biz `concat` function ishlatish uchun bank faylini ikkiga bo'lib olamiz

```python
bank1, bank2, bank3 = bank[:1000], bank[1000:2000], bank[2000:]
```

```python
bank1
```

```python
bank2
```

```python
bank3
```

```python
pd.concat([bank1, bank2, bank3]) # .concat() - ulash
```

```python
bank1 = bank1.set_index('age') # bank1 data'ni indexlarini ustunga utkazsak
bank1
```

#### Shunda bilib olishim kerakki qachonki biz ma'lumotlarni ulaganimizda, va indexlarini ustun nomida utkazganimizda, indexlar `unique` xususisiyatini yuqotadi

```python
bank2 = bank2.set_index('age')
bank2
```

```python
pd.concat([bank1, bank2], axis=1) 
```
