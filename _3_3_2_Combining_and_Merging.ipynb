##Sodda misollar

```python
A = pd.DataFrame({'Transport_turi':['poezd','avtomabil','trambvay'],
                  'Tezligi':[70, 180, 120]})
A
```

```python
B = pd.DataFrame({'Transport':['poezd','avtomabil','kema'],
                  'Narx':[89000, 120000, 44000]})
B
```

### `Merge`

```python
merged = pd.merge(A, B) # Inner join - ichki
merged
```

```python
merged_outer = pd.merge(A, B, how='outer') # outer join - tashqi
merged_outer
```

```python
merged_left = pd.merge(A, B, how='right') # right_join - chap
merged_left
```



##`Agar ustun nomlari bir xil bo'lsachi?`

```python
merg = pd.merge(A, B, left_on='Transport_turi',right_on='Transport') # ikki tarafgaham nom berib olamiz
merg 
```

## Indekslar ustida birlashtirish

```python
A = A.set_index('Transport_turi') # columns index
A
```

```python
merged1 = pd.merge(A, B, left_index=True, right_on='Transport')
```

```python
merged2 = pd.merge(A, B, left_index=True, right_index=True, how='outer')
merged2
```
